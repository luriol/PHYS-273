```python
from matplotlib import pyplot as plt # this imports the plotting program
import numpy as np
grades = np.array([61, 70, 80, 83, 90, 99])
hours = np.arange(1,7)
plt.plot(hours,grades,'ko')
fit_pars, cov = np.polyfit(hours, grades, 1, rcond=None, cov=True)
yfit = hours*fit_pars[0] + fit_pars[1]
plt.plot(hours,yfit,'-r')
plt.xlabel('Hours Spent Studying (hrs/week)')
plt.ylabel('Physics Grade')
plt.title('Physics Grade v. Number of Hourse of Study per Week')

uncert = (np.sqrt(np.diag(cov)))
plt.text(1.5,95,'y = ({0:4.1f} +-{2:3.1f}) x+({1:3.0f} +-{3:2.0f}) '.format(fit_pars[0],fit_pars[1],
              uncert[0],uncert[1]))
```
