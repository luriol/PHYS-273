This is an example of how to use python to do one of the webassign homework problems.  This example is particularly simple, I will try to provide one example per homework set and they will get more interesting as the semester goes on.

```python
nano = 1e-9 # define what a nano is
Q = 49*nano # charge of 49 nanocoulombs
# I am assuming MKS units throughout, so I only need to keep track of the nano part
q_e = 1.602e-19 #  definition of elementary charge (e.g. the charge of a proton)
N = Q/q_e
# the problem has now been solved.  All we need to do is print out the answer
print('The number of excess protons is {0:10.3e}'.format(N))
# The print syntax is to print everything between the quote marks.  The {0:7.3e} is a format specifier for the number N which we want to 
# print as a formatted string.  The "0" refers to the "first" argument of the format function.  (In python counting starts with zero). 
# the ":" seperates the argument from the format string.  The "7.3e" means print a number in exponential notation (the "e" part) using 10 total
# spaces with three of them after the decemil point.
```
